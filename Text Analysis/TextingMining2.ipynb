```python
#  Reading the text file AND appending to a list and analyzed to count the proportion of positive and negative words in the text
negativeWordlist=[]
with open("negative-words.txt","r") as negativeword:
    for words in negativeword:
        words=words.translate(words.maketrans("","",string.punctuation))
        words=words.split()
        for word in words:
            if word not in negativeWordlist:
                negativeWordlist.append(word)
    print(negativeWordlist)
    
                              
        
```

```python
#  Reading the text file and counting the number of times a word is being used and look for max and Minimum
from pathlib import Path
import os

path = Path(r"C:\Users\gw\inaugural\2017-Trump.txt")
print(path.exists())

# OR
worddic = {}
mypath = os.path.join(path, "/Users/gw/inaugural/", "2017-Trump.txt")
with open(mypath, "r") as file:
    for line in file:
        line = line.lower().split()
        for word in line:

            if word.isalpha():
                worddic[word] = 1+worddic.get(word, 0)
    print(worddic)
    
#  Getting the max and min words used in the file
max_word=""
max_count=0
for (word, count) in worddic.items():
    if count>max_count:
        max_word=word
        max_count=count
print(f" Most word used is: {max_word} count is: {max_count}")
```

```python
# Compute sentiments by looking at the proportion of positive and negative words in the obama text
from nltk import word_tokenize
cpositive=0
cnegative=0
for word in word_tokenize(obamaspeech):
    if word in positive_words:
        cpositive+=1
    if word in negativeWordlist:
        cnegative+=1
print(f'Obamaspeech postive words:{round((cpositive/len(word_tokenize(obamaspeech)))*100)}%,negative word:{round((cnegative/len(word_tokenize(obamaspeech)))*100)}%',"\t")
print("Obamaspeech positive words:{0:2.2%} and negative words {1:4.2%}\t".format(cpositive/len(word_tokenize(obamaspeech)),cnegative/len(word_tokenize(obamaspeech))))
```
