# <h2> Building the Model K-Nearest Neighours Classification/Regression</h2>
<p> The algorithm finds the point in the training set that is closet to the new point </p>
<p>Then it assigns the label of this new data point</p>
<p> k in the k-nearest neigbour signifies that instead of using only closest neighbour to the new data point, we consider
    we consider fixed number K of neigbours in the training data e.g three of five neigbours, and we make predictions using the majority class among the neigbour <p>
<p> Note Euclidean distance method and other methods are used for the algorithm</p>

# <h2> Every Model in sklearn is implemented in their own Classes,which are called Estimator Classes</h2>
<p>k-nearest neigbours classification algorithm is implemented in the KNeigbhoursClassifier class in the neighbors module</p>

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=20)  
# Note when the K is more than one the voting is done and the highest class in the neighbour is selected
# Note the objects knn encapsulate algorithm that will be used to build the model and predict the new data points
# because the KNeighborsClassifier holds only the training set
# but the knn object of the class KNeighborsClassifier will hold information that algorith
#has extracted from the training data
```

```python
knn.fit(X_train,y_train)
```

# <h4> We are investigating model complexity and generalization</h4>

```python
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 10
neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    # build the model
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # record training set accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    # record generalization accuracy
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

# <h2> Perform Features Engineering Otherwise know as interaction term</h2>
<li> to perform feature Engineering</li>
<li> Use <b> MinMaxScaler</b> for data standardization </li>
<li> from sklearn.preprocessing import MinMaxScaler</li>
<li> Use <b> PolynomialFeatures</b> for features engineering e,g  data  of [a,b] will return [1,a,b,a^2,ab,b^2] for degree 2 </li>
<li> for feature Engineering from sklearn.preprocessing import PolynomialFeatures</li>
<li> Remember the above are classes see implementation below</li>

```python
# This is the hard code for what was imported from  mglearn.datasets.load_extended_boston()
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PolynomialFeatures

def load_extended_boston():
    boston = load_boston()
    X=boston.data
    X=MinMaxScaler().fit_transform(boston.data)
    X=PolynomialFeatures(degree=2,include_bias=False).fit_transform(X)
    return X,boston.target
```
