```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv('Desktop\datasets\StudentsPerformance.csv')
print(data)
```

```python
data=data.dropna()
```

```python
data.describe()
```

```python
plt.scatter(data['math score'],data['reading score'], color='blue')
plt.title('Linear regression between math score and readings score',fontsize=14)
plt.xlabel('reading score',fontsize=14)
plt.ylabel('math score',fontsize=14)
plt.grid(False)
plt.show()
```

```python
# Multivvariated Regression : And this are the independent variable
X=data[['reading score','writing score']] #are the independent variables

# to display full statistical mode
X=sm.add_constant(X) #adding a constant
# This is dependebt variable
Y=data['math score']

model=sm.OLS(Y,X,missing="drop")
result=model.fit()
regression=result.summary()
print(regression)
```

```python
#  plotting fitted values with the actual Y values
fig,ax=plt.subplots(figsize=(8,6))
ax.plot(Y)
ax.plot(result.fittedvalues)
# The reason why the Graph looks distorted becuase there is no correlation between the data sets
```
