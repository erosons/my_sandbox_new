```python
amazon_data["Adj Close"].plot(kind="line",figsize=(13,7))

plt.ylabel("Amazon Close Price")
plt.xlabel("Year")

plt.show()
```

```python
monthlymean=amazon_data["Adj Close"].resample("M").mean()
monthlymean.head()
```

```python
monthlymedian=amazon_data["Adj Close"].resample("M").median()
monthlymedian.head()
```

```python
monthlymedian.plot(kind="area",figsize=(13,7))

plt.ylabel("Amazon Close Price")
plt.xlabel("Year")

plt.show()
```

```python
amazon_deltas=amazon_data["Adj Close"].diff()
amazon_deltas.head()
```

```python
amazon_returns=amazon_data["Adj Close"].pct_change()
amazon_returns.head(10)
```

```python
amazon_returns.plot(kind="line",figsize=(13,6),color="green")
plt.show()
```

```python
movingaverage_mean=amazon_returns.rolling(window=5).mean()
movingaverage_mean.head(10)
```

```python
movingaverage_mean.plot(kind="line",figsize=(13,6))
plt.show()
```

```python
fig = plt.figure(figsize=(13, 6))
original = plt.plot(amazon_returns, color='purple', label='Original') 
mean = plt.plot(movingaverage_mean, color='cyan', label='Rolling Mean')
plt.legend(loc='best') 
plt.title('Absolute Value and 5-day Rolling Mean of AMZN Returns') 
plt.show()
```
