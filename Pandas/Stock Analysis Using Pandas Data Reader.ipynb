```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas_datareader.data as data
from datetime import datetime
start = datetime(2001, 1, 1)
end=datetime.today()
findata=data.DataReader('BP', 'yahoo', start, end)
findata
```

```python
# single level filter or if logic
findata["Up"]=np.where(findata["Open"]-findata["Close"]>0,"up","down")
findata["Up"]
```

```python
#  plotting a bar chart
findata["Adj Close"].head(20).plot(kind="bar")
plt.show()
```

```python
#  plot the time series analysis
findata["Adj Close"].plot(kind="line",figsize=(13,7))

plt.ylabel("BP Adj Close Price")
plt.xlabel("Year")

plt.show()
```
